```python
import pandas as pd
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
df.columns=['age', 'gender', 'bmi', 'children', 'smoker', 'region', 'emi']
```

```python
df.head()
```

```python
pd.get_dummies(df)
```

```python
df1 = pd.get_dummies(df,drop_first="if_binary")     
df1.head()
```

```python
x = df1.drop('emi',axis=1)
y = df1[['emi']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2) 
```

```python
x_test
```

<h2>1. Linear Regression </h2>

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test['emi'])
```

```python
prediction = lr.predict(x_test)
```

```python
y_test['predicted_emi'] = prediction
```

```python
y_test
```

<h2> Decison Tree</h2>

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
print("Training score : ",round(dt.score(x_train,y_train),2)*100)  #when training is high and testing low, then it is overfit
print("Testing score : ",round(dt.score(x_test,y_test),2)*100)
```

<h2>Random Forest </h2>

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rdf = RandomForestRegressor()
```

```python
rdf.fit(x_train, y_train)
```

```python
print("Training score : ",round(rdf.score(x_train,y_train),2)*100)  #when training is high and testing low, then it is overfit
print("Testing score : ",round(rdf.score(x_test,y_test),2)*100)
```

```python
#random forest is best suited for this model
```

```python
import joblib
```

```python
model = 'Random_Forest.lb'
joblib.dump(rdf, model)
```

```python
loaded_model = joblib.load(model)
y_pred = loaded_model.predict(x_test)
print("Prediction: ", y_pred)
```
